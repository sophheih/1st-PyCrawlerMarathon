# 瀏覽器開發者工具介紹


* 知道瀏覽器的開發者工具
* 能夠使用瀏覽器的開發者工具觀察資料

* 元素面板（ Elements ）：顯示網頁中的 HTML 元素與佈局，可以在這裡做簡單的修改。另外注意右方會顯示選取的 HTML DOM 的 CSS 樣式。
* Console 可以用來執行 JavaScript ，通常我們會在上面做簡單的測試。
* Source 會顯示這次載入的網⾴頁所需要所有檔案，像是 CSS、JS 或是圖片等等。
* Network 會記錄所有網頁存取過程中的 HTTP 傳輸，每個HTTP 都會包含 Request 與 Response。
* Application 存放網頁在瀏覽器暫存的資料，包含 Cookie 或是 Storate 。
* 如何幫助爬蟲？
  Elements：觀察需要取得的 DOM 位置與 Selector 定位。
  Network：用來觀察存取資料的 API Request（尤其是在動態網⾴頁爬蟲）。
  Application：有一種登入機制會把權限狀態存在 Cookie ，可以在這裡查看。




## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```
